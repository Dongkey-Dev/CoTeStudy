```python
from tqdm import tqdm as tq

def have_pattern(S):
    result = ""
    for i in range(1, len(S)):
        if S[i-1] != S[i]:
            result += S[i-1]
    if 'A' in result[1:-1]:
        return False
    return True
    
def extract_header_0(S):
    result = ''
    S = S.replace('0.','')
    for i in S:
        if i == '0':
            result += i
        else:
            return result    
        
def reciprocal(N):
    div_result = extract_header_0(str(1/N))
    div_result += str(10**23750//N)
    
    for tail_element_index in range(1, len(div_result)):
        pattern = div_result[-tail_element_index:]
        replaced_by_pattern = div_result.replace(pattern, 'A')
        if have_pattern(replaced_by_pattern):
            min_pattern_start_index = len(div_result)
            for pos_pattern_comb in range(len(pattern)):
                start_point = 0
                combined_pattern = pattern[pos_pattern_comb:] + pattern[:pos_pattern_comb]
                if set(pattern) == {'0'}:
                    for k, element_div_result in enumerate(div_result):
                        if element_div_result != '0':
                            start_point = k
                            break
                combined_pattern_start_index = div_result[start_point:].index(combined_pattern) + start_point
                if min_pattern_start_index > combined_pattern_start_index:
                    pattern = combined_pattern
                    min_pattern_start_index = combined_pattern_start_index
            break
        
    if len(set(pattern))== 1:
        pattern = pattern[0]

    # print('0.' + div_result[:min_pattern_start_index] + pattern + ' ' + pattern)
    return ('0.' + div_result[:min_pattern_start_index] + pattern + ' ' + pattern)
    
res = []
for i in tq(range(100000, 2, -1)):
    try:
        reciprocal(i)
    except:
        res.append(i)

# reciprocal(99994)
```
